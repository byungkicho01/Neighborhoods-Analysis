# Comparable Neighborhoods

## Introduction/ Business Problem

There are many Canadians who move to the United States for the bigger job market and better opportunities. As of 2016, there were about 783k Canadians living in the U.S., and every year, about 50k Canadians immigrate to the U.S. While there are many similarities between the two countries, including language, education and geographical location, there are many challenges for many Canadians moving to the south of the border.One main challenge is in figuring out which area to live in.

For the purpose of this project, we will focus on neighborhoods in Downtown, Toronto and Manhattan, New York since there are many Torontonians looking to move to New York, especially in Finance, consulting and marketing. The problem this project will solve is: "What are the comparable neighborhoods between Downtown Toronto and Manhattan, NY?

This project will be beneficial to people who are looking to relocate between Downtown Toronto and Manhattan NY and looking for neighborhoods that offer similarities.

This project will extensively use Foursquare API though Rest API calls. Foursquare contains a extensive database of businesses globally including location, type of business, photos and user ratings. This data will be used in sync with the neighborhoods data in describing the distinct characteristics of neighborhoods in Toronto and NYC.

# Discussion
There are 8 distinctive clusters of neighborhoods in Downtown Toronto and Manhattan, NYC. Of the 8 clusters, 5 of them are clusters of neighborhoods from just Toronto or just NYC. However, there are 3 clusters that have mixes of neighborhoods from both Toronto and NYC. For example, if a person currently lives in University of Toronto, Harbord, Kensington Market, Chinatown, Grange Park, St. James Town or Cabbagetown in Toronto, he/she has 8 different neighborhoods in Manhattan, NY that are in the same cluster and have similar characteristics.
